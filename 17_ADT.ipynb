### Data type 

- defines memory storage 

- defines allowed operations 

##### - There are two types of data types :

   - 1. user defined data type
     
     - you can create your own data type containing your own value and function 
     - example - class , structure , union 
 
   - 2. built in data types 

### Absract data type

- ADT are like user defined data types and which defines operations on values using functions without specifying what is there inside a function and how opoerations are performed

## ðŸ§  Short Descriptions with Use Cases

### 1. **Stack (LIFO)**
- **Operations**: Push, Pop, Peek
- **Use Cases**: Undo feature, Function calls, Expression evaluation

### 2. **Queue (FIFO)**
- **Operations**: Enqueue, Dequeue, Front
- **Use Cases**: Task scheduling, Printer queues, BFS

### 3. **Deque (Double-Ended Queue)**
- **Operations**: Add/remove from both ends
- **Use Cases**: Sliding window problems, Palindrome checking

### 4. **List / Array / Linked List**
- **Operations**: Insert, Delete, Search
- **Use Cases**: General data storage, Navigation

### 5. **Set**
- **Operations**: Add, Remove, Membership test
- **Use Cases**: Removing duplicates, Set operations

### 6. **Map / Dictionary**
- **Operations**: Add key-value, Search by key
- **Use Cases**: Caching, Student marks database

### 7. **Tree (Binary Tree, BST, etc.)**
- **Operations**: Insert, Delete, Traverse
- **Use Cases**: Hierarchical data, Search systems

### 8. **Graph**
- **Operations**: Add edge, Remove edge, Traverse
- **Use Cases**: Social networks, Maps, Web crawling

### 9. **Priority Queue / Heap**
- **Operations**: Insert, Remove highest/lowest priority
- **Use Cases**: Dijkstraâ€™s algorithm, CPU scheduling
